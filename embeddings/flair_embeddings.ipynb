**Types**

```python
from flair.data import Sentence
```

```python
#slovenian
sentence = Sentence('Priročniki za učitelje ponujajo nasvete, kako čim bolje izkoristiti učbeniške vsebine, prinašajo tudi slikovno gradivo in dodatne vaje ter nasvete za realizacijo posameznih vsebin.')
sentence
```

```python
sentence.get_token(2)
```

```python
sentence[2]
```

```python
#Add custom tag
sentence[3].add_tag('ner', 'location')
sentence.to_tagged_string()
```

**Word embedding**

```python
from flair.embeddings import WordEmbeddings
```

Classical embeddings

```python
# init embedding
glove_embedding = WordEmbeddings('glove') #100dim
# embed a sentence using glove.
glove_embedding.embed(sentence)

# now check out the embedded tokens.
for token in sentence:
    print(token)
    print(token.embedding)
```

```python
# init embedding
fasttext_embedding = WordEmbeddings('sl-crawl') #300dim
# embed a sentence using FastText.
fasttext_embedding.embed(sentence)

# now check out the embedded tokens.
for token in sentence:
    print(token)
    print(token.embedding)
```

Flair embeddings

```python
from flair.embeddings import FlairEmbeddings

# init embedding
# flair_embedding_forward = FlairEmbeddings('news-forward')
flair_embedding_forward = FlairEmbeddings('sl-forward')
flair_embedding_backward = FlairEmbeddings('sl-backward')

# embed words in sentence
flair_embedding_forward.embed(sentence)
```

Stacked embeddings

```python
from flair.embeddings import WordEmbeddings, CharacterEmbeddings, FlairEmbeddings, StackedEmbeddings
```

```python
# init standard GloVe embedding
fasttext_embedding = WordEmbeddings('sl-crawl')

# init Flair forward and backwards embeddings
flair_embedding_forward = FlairEmbeddings('sl-forward')
flair_embedding_backward = FlairEmbeddings('sl-backward')

# create a StackedEmbedding object that combines glove and forward/backward flair embeddings
stacked_embeddings = StackedEmbeddings([
                                        fasttext_embedding,
                                        flair_embedding_forward,
                                        flair_embedding_backward,
                                       ])
# just embed a sentence using the StackedEmbedding as you would with any single embedding.
stacked_embeddings.embed(sentence)

# now check out the embedded tokens.
for token in sentence:
    print(token)
    print(token.embedding)
    
# Words are now embedded using a concatenation of three different embeddings. 
# This means that the resulting embedding vector is still a single PyTorch vector.
```
